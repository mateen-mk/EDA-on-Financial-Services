```python
# import necessory libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import data set using pandas dataframe
df = pd.read_csv('application_data.csv')
```

```python
# for displaying all the columns in upcoming steps
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows', None)
```

```python
# create point plot to show the percentage of null values in data and draw a line on 45%
null_data = pd.DataFrame(df.isnull().mean()*100).reset_index()
null_data.columns = ['Columns','Percentage of Null values']
plt.figure(figsize = (28,6))
sns.pointplot(x = 'Columns', y = 'Percentage of Null values', data = null_data, color= 'orange').axhline(45,ls ='--', color = 'green')
plt.xticks(rotation = 90, fontsize = 7)
plt.title('Percentage of Null values in Financial Services data')
plt.xlabel('Columns')
plt.ylabel('Percentage of Null values')
plt.show()
```

```python
# storing all the columns which have null values more then 50% in new table
null_df = null_data.sort_values('Percentage of Null values',ascending= False).head(49).reset_index(drop = True)
null_df.drop('Percentage of Null values', axis = 1, inplace= True)
```

```python
# columns which are matching index numbers --> 6, 40, 30, 23
null_df = null_df.drop([6,23,30,40], axis= 0)
null_list = null_df['Columns'].tolist()
```

```python
# finding the percentage of null values -- if the percentage of null values is more then 50% and that is not relavent then drop the entire column
df.drop(columns=null_list, inplace= True)
```

```python
df.drop(columns='EXT_SOURCE_1', inplace= True)
df.drop(columns='EXT_SOURCE_2', inplace= True)
df.drop(columns='EXT_SOURCE_3', inplace= True)
df.drop(columns='HOUSETYPE_MODE', inplace= True)
df.drop(columns='FONDKAPREMONT_MODE', inplace= True)
df.drop(columns='WALLSMATERIAL_MODE', inplace= True)
```

```python
# Handle Null values of object type data
df['OCCUPATION_TYPE'] = df['OCCUPATION_TYPE'].fillna('Not Working')
df['NAME_TYPE_SUITE'].fillna(value=df['NAME_TYPE_SUITE'].value_counts().index[0], inplace=True)
```

```python
# Handle Null values of float type data
float_cols = df.select_dtypes(include=['float']).columns
df[float_cols] = df[float_cols].fillna(df[float_cols].mean())
```

```python
# Define styles
plt.style.use('dark_background')
sns.set_style('darkgrid')
```

```python
# Loan Approval Rate
approved_rate = df['TARGET'].mean()*100
not_approved_rate = 100-approved_rate
labels = ['Not Approved', 'Approved']
sizes = [approved_rate, not_approved_rate]
colors = ['orange', 'lightgreen']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Loan Approval Rate')
plt.show()
```

```python
# Average Income
sns.set(style='darkgrid')
plt.figure(figsize=(16, 4))
sns.boxplot(data=df, x='AMT_INCOME_TOTAL')
plt.title('Distribution of Average Income')
plt.xlabel('Average Income')
plt.show()
```

```python
# Credit Amount vs. Income Ratio
df['credit_income_ratio'] = df['AMT_CREDIT'] / df['AMT_INCOME_TOTAL']
plt.figure(figsize = (16,4))
sns.histplot(df['credit_income_ratio'], kde=True, color='#ff9999')
plt.title('Credit Amount vs. Income Ratio')
plt.xlabel('Credit/Income Ratio')
plt.ylabel('Count')
plt.show()
```

```python
# Age of Applicants
df['age'] = abs(df['DAYS_BIRTH']) // 365
plt.figure(figsize = (16, 5))
sns.histplot(df['age'], kde=True, color='purple')
plt.title('Age of Applicants')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python

# group days employed into categories
df['Employment History'] = pd.cut(df['DAYS_EMPLOYED']/-365, bins=[-np.inf, 1, 4, 7, 12, np.inf], labels=['< 1 year', '1-4 years', '4-7 years', '7-12 years', '> 12 years'])
history_counts = df.groupby('Employment History').size().reset_index(name='counts')

# create a pie chart for employment history
fig, ax = plt.subplots(figsize=(6, 6))
labels = history_counts['Employment History']
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#f0e68c']
explode = [0.05, 0.05, 0.05, 0.05, 0.05]
ax.pie(history_counts['counts'], colors=colors, labels=labels, explode=explode, autopct='%1.1f%%', startangle=90)
plt.legend(loc='upper right')
plt.title('Employment History')
plt.show()
```

```python
# Converting data type to integer
df['FLAG_OWN_CAR'] = df['FLAG_OWN_CAR'].replace({'Y': 1, 'N':0})
df['FLAG_OWN_REALTY'] = df['FLAG_OWN_REALTY'].replace({'Y': 1, 'N': 0})

# Ownership Percentage
car_owners = df['FLAG_OWN_CAR'].mean()*100
realty_owners = df['FLAG_OWN_REALTY'].mean()*100
labels = ['Car Owners', 'Realty Owners']
sizes = [car_owners, realty_owners]
colors = ['purple', 'skyblue']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Ownership Percentage')
plt.show()
```

```python
# Distribution of Education Types
plt.figure(figsize=(12,5))
sns.countplot(y='NAME_EDUCATION_TYPE', data=df, color='purple')
plt.title('Distribution of Education Types')
plt.xlabel('Count')
plt.ylabel('Education Type')
plt.show()
```

```python
# Distribution of Family Status
plt.figure(figsize=(12,5))
sns.countplot(y='NAME_FAMILY_STATUS', data=df, color='#ff9999')
plt.title('Distribution of Family Status')
plt.xlabel('Count')
plt.ylabel('Family Status')
plt.show()
```

```python
# Distribution of Housing Types
plt.figure(figsize=(12,5))
sns.countplot(y='NAME_HOUSING_TYPE', data=df, color='brown')
plt.title('Distribution of Housing Types')
plt.xlabel('Count')
plt.ylabel('Housing Type')
plt.show()
```

```python
# credit bureau inquiries
plt.figure(figsize=(10, 6))
sns.barplot(x=['Hour', 'Day', 'Week', 'Month', 'Quarter', 'Year'], 
            y=[df['AMT_REQ_CREDIT_BUREAU_HOUR'].mean(), df['AMT_REQ_CREDIT_BUREAU_DAY'].mean(), 
               df['AMT_REQ_CREDIT_BUREAU_WEEK'].mean(), df['AMT_REQ_CREDIT_BUREAU_MON'].mean(), 
               df['AMT_REQ_CREDIT_BUREAU_QRT'].mean(), df['AMT_REQ_CREDIT_BUREAU_YEAR'].mean()],
            palette='Set2')

plt.title('Average Credit Bureau Inquiries')
plt.xlabel('Time Period')
plt.ylabel('Number of Inquiries')
plt.gca().yaxis.set_major_formatter('{:,}'.format)
plt.show()
```
